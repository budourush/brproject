ある数 x が x = a^y で表されているときの指数 y を、a を底とする x の対数といい、記号 log を用いて

真数を変数とした関数が対数関数。

真数は必ず x > 0 となる。

```python
x = np.linspace(0.00001, 10, 1000)

y1 = np.log2(x)
y2 = np.log10(x)

fig, (axL, axR) = plt.subplots(ncols=2, figsize=(10,4))

axL.plot(x, y1, linewidth=2)
axL.set_title('log2(x)')
axL.set_ylabel('x')
axL.set_ylim(-6, 4)
axL.grid(True)

axR.plot(x, y2, linewidth=2)
axR.set_title('log10(x)')
axR.set_xlabel('t')
axR.set_ylabel('x')
axR.set_ylim(-6, 4)
axR.grid(True)

fig.show()
```

<div style="text-align: center;">
対数に関する公式
</div>

$$
\begin{eqnarray}
    &①& \log_{ a } a = 1\\
    &➁& \log_{ a } 1 = 0\\
    &③& \log_{ a } XY = \log_{ a } X + \log_{ a } Y\\
    &④& \log_{ a } \frac{ X }{ Y } = \log_{ a } X - \log_{ a } Y\\
    &⑤& \log_{ a } X^p = p * log_{ a } X\\
    &⑥& \log_{ a } X = \frac{ log_{ c } X }{ log_{ c } a }\\
\end{eqnarray}
$$

公式について考える
$$
    log_{ a } a = 1\\
    log_{ a } 1 = 0\\
$$
は対数の定義よりわかる。

---------------------

$$
    \log_{ a } XY = \log_{ a } X + \log_{ a } Y\\
$$
を考える。
$$
    \log_{ a } X = u_x\\
    \log_{ a } Y = u_y
$$
とすると
$$
    X=a^{u_x}\\
    Y=a^{u_y}
$$
となり
$$
    XY = a^{u_x} * a^{u_y} = a^{u_x + u_y}
$$
ここで対数の定義を考えると
$$
    \log_{ a } XY = u_x + u_y
$$
よって
$$
    \log_{ a } XY = \log_{ a } X + \log_{ a } Y
$$

---------------------

$$
    \log_{ a } X^p = p * log_{ a } X\\
$$
を考える。
$$
    \log_{ a } X = u\\
$$
とすると
$$
    X=a^u
$$
となり
$$
    X^p = {(a^u)}^p = a^{(pu)}
$$
ここで対数の定義を考えると
$$
    \log_{ a } {X^p} = pu
$$
よって
$$
    \log_{ a } X^p = p * log_{ a } X
$$
真数の指数はlog自体に掛け算できる。

---------------------

$$
    \log_{ a } \frac{ X }{ Y } = \log_{ a } X - \log_{ a } Y\\
$$
を考える。
$$
    \log_{ a } \frac{ X }{ Y } = \log_{ a } {XY^{-1}}
$$
とすると
$$
    \log_{ a } {XY^{-1}} = \log_{ a } X + \log_{ a } Y^{-1}
$$
となり、ここで
$$
    \log_{ a } Y^{-1} = -\log_{ a } Y
$$
なので
$$
    \log_{ a } \frac{ X }{ Y } = \log_{ a } X - \log_{ a } Y
$$
が証明できる。

<div style="text-align: center;">
ネイピア数と極限
</div>

$$
    \lim_{ n \to \infty } (1 + \frac{ 1 }{ n })^n = e(ネイピア数)
$$
から
$$
    n = \frac{1}{t}
$$
とすると
$$
    \lim_{ n \to \infty }n = \lim_{ t \to 0 }\frac{1}{t} = \infty
$$
となるため、ネイピア数の公式は
$$
    \lim_{ t \to 0 }(1+t)^{\frac{1}{t}} = e
$$
とも表現できる。

---------------

$$
    \lim_{ t \to 0 }(1 + t)^{\frac{ 1 }{ t } } = e
$$
の両辺から対数をとると
$$
\begin{eqnarray}
    \lim_{ t \to 0 }\log{(1 + t)^{\frac{ 1 }{ t } }}&=&\log{e}\\
    \lim_{ t \to 0 }{\frac{ \log{(1+t)} }{ t }} &=& 1
\end{eqnarray}
$$
を導ける。

---------------

$$
    \lim_{ t \to 0 }{\frac{ \log{(1+t)} }{ t }} = 1
$$
において
$$
    t = e^x -1 
$$
とすると。
$$
    \lim_{ t \to 0 }t = \lim_{ x \to 0 }e^x -1 = 0
$$
なので
$$
\begin{eqnarray}
    \lim_{ x \to 0 }{\frac{ \log{e^x} }{ e^x -1 }} &=& 1\\
    \lim_{ x \to 0 }{\frac{ x\log{e} }{ e^x -1 }} &=& 1\\
    \lim_{ x \to 0 }{\frac{ x }{ e^x -1 }} &=& 1\\
\end{eqnarray}
$$
逆数をとると
$$
    \lim_{ x \to 0 }{\frac{ e^x -1 }{ x }} = 1
$$
を導ける。

関数f(x)の2点((x , f(x)) , (x+h , f(x+h)))を考える。
hを0に近づけることでf(x)の点xにおける傾きを求めることができる。

傾き = (yの増加量) / (xの増加量)
$$
    \frac{ f(x+h) - f(x)}{ (x+h) - x}\\
$$

点xの傾きaを求めるにはhを0に近づけることで求めることができる。
$$
    a = \lim_{ h \to 0 } \frac{ f(x+h) - f(x)}{ (x+h) - x}\\
$$

上記の式をf(x)の導関数といい
$$
    f'(x)\quad(x)'\quad\frac{ dy }{ dx }
$$
で表す。

よって
$$
    f'(x) = \lim_{ h \to 0 } \frac{ f(x+h) - f(x)}{ (x+h) - x}\\
$$
となる。

ここで 
$$
f(x) = 2x^2 
$$
とすると

$$
\begin{eqnarray}
    f'(x) &=& \lim_{ h \to 0 } \frac{ 2(x+h)^2 - 2x^2}{ (x+h)- x}\\
    f'(x) &=& \lim_{ h \to 0 } \frac{ 2(x^2+2xh+h^2) - 2x^2}{ h }\\
    f'(x) &=& \lim_{ h \to 0 } \frac{ 2x^2+4xh+2h^2 - 2x^2}{ h }\\
    f'(x) &=& \lim_{ h \to 0 } \frac{ 4xh+2h^2 }{ h }\\
    f'(x) &=& \lim_{ h \to 0 } 4x+2h\\
    f'(x) &=& 4x
\end{eqnarray}
$$
となる。

接線の傾きは<br>
x = 5 の時
$$
    f'(5) = 4 * 5 = 20
$$

x = 10 の時

$$
    f'(10) = 4 * 10 = 40
$$
となる。

    &⑤&f(x)*g(x) = f'(x)g(x) + f(x)g'(x)\\

\end{eqnarray}
$$
④微分の和

に関しては導関数を求める計算を便利にするための公式。

---------------

$$
    (\log{ x })' = \frac{ 1 }{ x }\\
$$
を証明する。

微分の定義式に当てはめると
$$
\begin{eqnarray}
    (\log{ x })' &=& \lim_{h \to 0}\frac{ \log{(x+h)} - \log{(x)}}{ h }\\
    &=&\lim_{h \to 0}\frac{1}{h}\log{\frac{x+h}{x}}\\
    &=&\lim_{h \to 0}\frac{x}{h}\times\frac{1}{x}\log{(1+{\frac{h}{x})}}\\
\end{eqnarray}
$$
ここで
$$
    \frac{h}{x} = k
$$
とすると
$$
    h \to 0 = k \to 0であるから\\
\begin{eqnarray}
    (\log{ x })' &=& \lim_{k \to 0}\frac{1}{x}\times\frac{1}{k}\log{(1+k)}\\
    &=&\frac{1}{x}\lim_{k \to 0}\log{(1+k)^{\frac{1}{k}}}
\end{eqnarray}
$$
ここで
$$
    \lim_{ t \to 0 }(1+t)^{\frac{1}{t}} = e
$$
なので
$$
\begin{eqnarray}
    (\log{ x })' &=&\frac{1}{x}\log{e}\\
    &=&\frac{1}{x}
\end{eqnarray}
$$
が導ける。


---------------

$$
    (e^x)' = e^x\\
$$
を証明する。

微分の公式に当てはめると
$$
\begin{eqnarray}
    (e^x)'&=&\lim_{ h \to 0 }\frac{e^{x+h}-e^x}{h}\\
    &=&\lim_{ h \to 0 }\frac{e^xe^h-e^x}{h}\\
    &=&\lim_{ h \to 0 }\frac{e^x(e^h-1)}{h}\\
    &=&e^x\lim_{ h \to 0 }\frac{(e^h-1)}{h}\\
    \lim_{ h \to 0 }\frac{(e^h-1)}{h}&=&1より\\
    (e^x)'&=&e^x\\
\end{eqnarray}
$$

---------------

$$
    (x^n)' = nx^{n-1}\\
$$
を証明する。
$$
    f(x) = x^n
$$
とすると、両辺の対数とり、xで微分すると
$$
\begin{eqnarray}
    \log{f(x)} &=& n\log{x}\\
    \frac{f'(x)}{f(x)} &=& \frac{n}{x}\\
    f'(x) &=& f(x)\frac{n}{x}\\
    f'(x) &=& n\frac{x^n}{x}\\
    f'(x) &=& nx^{n-1}
\end{eqnarray}
$$
が導ける。

    f(x) =  2x^2 - 5x + 11
$$
の時f'(x)を求める。


公式①④を利用する。
$$
\begin{eqnarray}
    g(x) &=& 2x^2\\
    h(x) &=& -5x\\
    j(x) &=& 11\\
\end{eqnarray}
$$
とすると
$$
    f(x) = g(x) + h(x) + j(x)
$$
ここで公式④を利用すると
$$
    f'(x) = g'(x) + h'(x) + j'(x)
$$

となり後はg'(x) + h'(x) + j'(x)を求めればf'(x)が求まる。ここで公式①を利用すると

$$
\begin{eqnarray}
    g'(x) &=& 4x\\
    h'(x) &=& -5\\
    j'(x) &=& 0\\
\end{eqnarray}
$$
よって
$$
    f'(x) = 4x -5
$$
