対数問題回答

$$
\begin{eqnarray}
    3\log_{3}3 + \log_{3}\frac{3}{4} + \log_{3}12 &=& 3 \times 1 + \log_{3}(\frac{3}{4}\times12)\\
    &=& 3 + \log_{3}9\\
    &=& 3 + \log_{3}3^2\\
    &=& 3 + 2\log_{3}3\\
    &=& 3 + 2 \times 1\\
    &=& 3 + 2\\
    &=& 5
\end{eqnarray}
$$

$$
\begin{eqnarray}
    4\log_{2}\sqrt{6} - 2\log_{2}3 &=& 4\log_{2}6^{\frac{1}{2}} - \log_{2}{3}^2\\
    &=& \log_{2}6^{(\frac{1}{2}\times 4)} - \log_{2}{3}^2\\
    &=& \log_{2}6^2 - \log_{2}{3}^2\\
    &=& \log_{2}\frac{6^2}{3^2}\\
    &=& \log_{2}\frac{36}{9}\\
    &=& \log_{2}4\\
    &=& \log_{2}2^2\\
    &=& 2\log_{2}2\\
    &=& 2
\end{eqnarray}
$$

$$

$$

f(2)の接線と傾きを求める。

----------

まずf(2)の時の値を計算する。
$$
\begin{eqnarray}
    f(2) &=& 2^3 + 3 \times 2^2 - 9 \times 2 + 4\\
    &=& 8 + 12 - 18 + 4\\
    &=& 6
\end{eqnarray}
$$

f(2) = 6 であるので接線は点(2,6)を通る。

----------

次に接線の傾きを求める。
$$
    f(x) = x^3 + 3x^2 - 9x + 4
$$
より、微分の和の公式を利用して導関数を求める。
$$
    f'(x) = 3x^2 + 6x - 9
$$
よってf(2)の時の接線は
$$
\begin{eqnarray}
    f'(2) &=& 3 \times 2^2 + 6 \times 2 -9\\
    &=& 12 + 12 -9\\
    &=& 15
\end{eqnarray}
$$
